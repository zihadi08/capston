```python

import os
import cv2
import pandas as pd
import matplotlib.pyplot as plt
import easyocr
from pathlib import Path

# Initialize OCR
ocr = easyocr.Reader(['bn'])

# Input/output paths
input_dir = "/kaggle/input/yolo11-test-images"
output_dir = "/kaggle/working/output-images/"
csv_path = "license_results.csv"

os.makedirs(output_dir, exist_ok=True)

def process_image(image_path):
    img = cv2.imread(image_path)
    if img is None:
        print(f"⚠️ Could not read {image_path}")
        return None, None
    
    # Run OCR
    results = ocr.readtext(image_path)

    # Collect detected texts
    detected_texts = [line[1] for line in results]
    plate_number = "".join(detected_texts).strip()
    filename = Path(image_path).name

    # Create side-by-side visualization
    fig, ax = plt.subplots(1, 2, figsize=(10, 4))
    
    fig.suptitle(f"File: {filename}", fontsize=14, weight="bold")
    ax[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    ax[0].set_title("Original")
    ax[0].axis("off")

    ax[1].text(0.5, 0.5, plate_number if plate_number else "N/A",
               fontsize=20, ha="center", va="center")
    ax[1].set_title("Predicted Plate")
    ax[1].axis("off")

    output_file = Path(output_dir) / (Path(image_path).stem + "_result.png")
    plt.savefig(output_file, bbox_inches="tight")
    plt.show()
    plt.close()

    return Path(image_path).name, plate_number

# Process all images
results_data = []
for img_file in Path(input_dir).glob("*.*"):
    fname, number = process_image(str(img_file))
    if fname and number:
        results_data.append({"filename": fname, "plate_number": number})

# Save results to CSV
df = pd.DataFrame(results_data)
df.to_csv(csv_path, index=False)

print(f"✅ Done! Results saved to {csv_path} and visualizations in {output_dir}")
```
